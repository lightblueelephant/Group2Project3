```python
# Continue preparing the data to fit a predictive model on cases either "Paid" or "Unpaid"

paid = concat_df[concat_df["Status"] == "Paid"]
unpaid = concat_df[concat_df["Status"] == "Unpaid"]
paid_unpaid_df = pd.concat([paid, unpaid])
paid_unpaid_df.sort_index(inplace=True)
paid_unpaid_df
```

```python
# create a target column "Paid" which reflects whether a case has paid (1) or unpaid (0) status

paid_unpaid_df['Paid'] = np.nan
        
for index, row in paid_unpaid_df.iterrows():
    if row["Status"] == "Paid":
        paid_unpaid_df.loc[index, "Paid"] = 1
    else:
        paid_unpaid_df.loc[index, "Paid"] = 0

paid_unpaid_df = paid_unpaid_df.drop(columns="Status")
paid_unpaid_df[0:100]
```

```python
state_dummies = pd.get_dummies(paid_unpaid_df['State'])
state_dummies
```

```python
make_dummies = pd.get_dummies(paid_unpaid_df['Vehicle Make'])
make_dummies.head()
```

```python
color_dummies = pd.get_dummies(paid_unpaid_df['Vehicle Color'])
color_dummies.head()
```

```python
vehicle_df = pd.concat([state_dummies, make_dummies, color_dummies], axis=1)
vehicle_df
```

```python
parking_details_df = paid_unpaid_df[['Zone', 'Base Amount', 'Penalty Fee']]
X = pd.concat([vehicle_df, parking_details_df], axis=1)
X = X.drop(columns=0) 
X
```

```python
y = paid_unpaid_df['Paid']
y
```

```python
# import the LogisticRegression model, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
lr_model = LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)

lr_model.fit(X_train, y_train)
```

```python
predictions = lr_model.predict(X_train)
predictions_df = pd.DataFrame({'Predictions':predictions,'Actual':y_train})
predictions_df[0:100]
```
